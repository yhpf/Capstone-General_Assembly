### Function to get lyrics for specific song and artist

```python
def get_lyrics(song, artist):
    response = api.search_song(song, artist)
    if response:
        return response.lyrics
    else:
        return None
```

## Load the data

```python
df = pd.read_csv('./songs_lyrics_api_data.csv')
```

## Look at the data

```python
df.head(3)
```

#### Make a data frame with only unique songs, so songs don't repeat

```python
df2 = df.drop_duplicates(subset='ID')
```

```python
df2.shape
```

#### Make new data frame to make sure that I am only get lyrics for the songs that I am missing lyrics for

```python
df3 = df2[df2.isnull()['Lyrics']]
```

```python
df3.shape
```

#### Make new data frame for the choosen 8 music industry markets

Originaly 10 music markets but 2 asian markets were excluded due to asian sign in the lyrics.

```python
df4 = df3[(df3['Country'] == 'us') | (df3['Country'] == 'gb') | (df3['Country'] == 'de') | (df3['Country'] == 'fr') |
         (df3['Country'] == 'ca') | (df3['Country'] == 'au') | (df3['Country'] == 'it') | (df3['Country'] == 'nl')]
```

```python
df4.shape
```

#### Make list of lists

```python
l = df4
n = 50

lists = [l[i:i + n] for i in range(0, len(l), n)]
```

```python
lists[0]
```

#### Make a for-loop to get lyrics based on the lists

Include to print something indicating the process.

```python
for i, lst in enumerate(lists):
    print('### scraping list ', i)
    for j, row in lst.iterrows():
        df4.loc[j,'Lyrics'] = get_lyrics(row['Track Name'], row['Artist'])
```
