### What does it look like for the 8 countries each?

```python
def country_null(country):
    miss = data[data['Country'] == country].isnull().sum()['Lyrics']
    total = data[data['Country'] == country].shape[0]
    return miss / total
```

```python
[(c, country_null(c)) for c in COUNTRIES]
```

#### Total lyrics rate per country

```python
[(c, 1 - country_null(c)) for c in COUNTRIES]
```

```python
COUNTRIES = ['us', 'gb', 'de', 'fr', 'ca', 'au', 'it', 'nl']
```

#### Function to get share of missing data

```python
def country_null3(country):
    miss = df2[df2['Country'] == country].isnull().sum()['Lyrics']
    total = df2[df2['Country'] == country].shape[0]
    return miss / total
```

#### Total missing data rate per market

```python
[(c, country_null3(c)) for c in COUNTRIES]
```

#### Total lyrics rate per country

```python
[(c, 1 - country_null3(c)) for c in COUNTRIES]
```
