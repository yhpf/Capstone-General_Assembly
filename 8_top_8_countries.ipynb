#### What does it look like for the 8 countries each?

```python
COUNTRIES = ['us', 'gb', 'de', 'fr', 'ca', 'au', 'it', 'nl']
```

```python
def country_null(country):
    miss = data[data['Country'] == country].isnull().sum()['Lyrics']
    total = data[data['Country'] == country].shape[0]
    return miss / total
```

```python
# total missing data rate per country
[(c, country_null(c)) for c in COUNTRIES]
```

```python
# total lyrics rate per country
[(c, 1 - country_null(c)) for c in COUNTRIES]
```
