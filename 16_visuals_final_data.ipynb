# Visuals on the final data

## Import stuff

```python
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use('seaborn-bright')

%matplotlib inline
```

## Load data

```python
big_data = pd.read_csv('./big_df_song20180401.csv')
big_data.head(3)
```

```python
cvec_train_data = pd.read_csv('./final_train_cvec.csv')
cvec_train_data.head(3)
```

```python
tvec_train_data = pd.read_csv('./final_train_tvec.csv')
tvec_train_data.head(3)
```

## Make some plots

### Plot for the big data set

```python
big_data.head(3)
```

```python
big_data1 = big_data.loc[:, ['Track Name', 'Artist', 'ID', 'Lyrics', 'Acousticness',
       'Energy', 'Instrumentalness', 'Mode', 'Tempo', 'Valence', 'au', 'ca',
       'de', 'fr', 'gb', 'it', 'nl', 'us', 'avg_Streams', 'avg_Position',
       'Polarity', 'Subjectivity']]
```

```python
big_data1.columns
```

```python
plt.figure(figsize=(9,6))
plt.scatter(big_data1['Energy'], big_data1['Valence'], color='blue')
plt.plot([0,1],[0,1], c='r')
#plt.title('Energy vs. Target', fontsize=18)
plt.xlabel('Energy', fontsize=14)
plt.ylabel('Valence (Target)', fontsize=14);

plt.rc('axes', titlesize=16)     
plt.rc('axes', labelsize=14)    
plt.rc('xtick', labelsize=14)   
plt.rc('ytick', labelsize=14)
plt.rc('figure', titlesize=14);
```

```python
big_data1['Energy'].plot(kind='hist', figsize=(5,5), color='red'); 
plt.xlabel('Energy');
plt.ylabel('Number of tracks');
plt.title('Energy');
```

```python
big_data1['Acousticness'].plot(kind='hist', figsize=(5,5), color='red'); 
plt.xlabel('Acousticness');
plt.ylabel('Number of tracks');
plt.title('Acousticness');
```

```python
big_data1['Polarity'].plot(kind='hist', figsize=(5,5), color='red'); 
plt.xlabel('Polarity = neg (-1) and pos (1)');
plt.ylabel('Number of tracks');
plt.title('Polarity');
```
